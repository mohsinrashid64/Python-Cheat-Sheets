## Hierarchical Indexing

```python
A = [['Bulb','Bulb','Bulb','Fan','Fan','Fan',],
     ['A','B','C','A','B','C']
                  ]
indx = pd.MultiIndex.from_arrays(A,names=('ProductName','Type'))
df = pd.DataFrame({'EC':[20,30,40,25,10,30]},index=indx)
```
