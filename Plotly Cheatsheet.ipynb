## Simple Plot

```python
import plotly.graph_objs as go
Z=np.linspace(0,10,50)
X=np.cos(Z)
Y=np.sin(Z)
trace=go.Scatter3d(x=X,y=Y,z=Z,mode="markers",marker=dict(size=12,color=Z,colorscale='Viridis'))
layout=go.Layout(title='3D interactive plot')
fig=go.Figure(data=[trace],layout=layout)
fig.show()
```

## Surface Plot

```python
X,Y=np.meshgrid(np.linspace(-2,2,30),np.linspace(-2,2,30))
Z=np.cos(X**2+Y**2)
trace=go.Surface(x=X,y=Y,z=Z)
fig=go.Figure(data=[trace],layout=go.Layout(title='3D Surface'))
# go.Figure(data=[trace],layout=go.Layout(title='3D Surface')).show this can be done too

fig.show()
```
