## Setting Figure Size, Colors and Styles of the Lines

```python
X=np.linspace(0,10,1000)

fig=plt.figure(figsize=(10,10))
ax=plt.axes()

ax.plot(X,np.sin(X-0),color='blue')
ax.plot(X,np.sin(X-1),color='g')
ax.plot(X,np.sin(X-2),color=(1.0,0.2,0.3))

plt.plot(X,X+0,linestyle='solid')
plt.plot(X,X+1,linestyle='dashed')
plt.plot(X,X+2,linestyle='dashdot')
plt.plot(X,X+3,linestyle='dotted')
```

## Shortcuts For Colors and Styles

```python
X=np.linspace(0,10,1000)

fig=plt.figure(figsize=(10,10))
ax=plt.axes()

ax.plot(X,np.sin(X-0),linestyle='--') # dashed
ax.plot(X,np.sin(X-1),linestyle='-') # solid
ax.plot(X,np.sin(X-2),linestyle=':') # dotted

plt.plot(X,X+0,'-g') # solid green
plt.plot(X,X+1,'--c') # dashed cyan
plt.plot(X,X+2,'-.k') # dashdot black
plt.plot(X,X+3,':r') # dotted red
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)

fig = plt.figure(layout="constrained")

gs = GridSpec(3, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
# identical to ax1 = plt.subplot(gs.new_subplotspec((0, 0), colspan=3))
ax2 = fig.add_subplot(gs[1, :-1])
ax3 = fig.add_subplot(gs[1:, -1])
ax4 = fig.add_subplot(gs[-1, 0])
ax5 = fig.add_subplot(gs[-1, -2])

fig.suptitle("GridSpec")
format_axes(fig)

plt.show()
```

## Slider

```python
from matplotlib.widgets import Slider
A = pd.Series([2,3,4,5],index=['Issue Period','Issued','Denied','Pending'])
data = {
     'Issued':[48599,22815,2141],
     'Denied':[1733,400,323],
     'Pending':[336,1013,2701]
}
```

```python
df = pd.Series(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
periods = [str(_) for _ in df.index]
issued = df['Issued']
denied = df['Denied']
pending = df['Pending']
dataset = [issued,denied,pending]
```

```python
%matplotlib notebook
fig,ax = plt.subplots(figsize=(9,6))
ax.plot(periods,issued)
plt.subplots_adjust(bottom=0.25)
ax_slider = plt.axes([0.1,0.1,0.8,0.05])


def update_line(indx):
    ax.clear()
    ax.plot(periods,dataset[indx])
    ax.set_title(dataset[indx].name)
    plt.draw()
    

slider = Slider(ax_slider, "Category",valmin=0,valmax=2,valinit=0,valstep=1)
slider.on_changed(update_line)

plt.show()
```

```python
%matplotlib inline
from scipy import *
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

## Definition of the plot_cos function, our "callback function".
def plot_cos(phi):
    ## Plot parameters
    xmin, xmax, nx = 0.0, 10.0, 50
    ymin, ymax     = -1.2, 1.2

    ## Plot the figure
    x = np.linspace(xmin, xmax, nx)
    y = np.cos(x + phi)
    plt.figure(figsize=(8,3))
    plt.plot(x, y, linewidth=2)

    ## Set up the figure axes, etc.
    plt.title("y = cos(x + phi)")
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.xlabel('x')
    plt.ylabel('y')

## Generate our user interface.
interact(plot_cos, phi=FloatSlider(min=-3.2, max=3.2, step=0.2, value=0.0));
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Create some data
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the initial data
line, = ax.plot(x, y, 'b')

# Set the initial slider value
initial_value = 1.0

# Create a slider axes
slider_ax = plt.axes([0.1, 0.05, 0.8, 0.03])

# Create a slider
slider = Slider(slider_ax, 'Slider', 1.0, 10.0, valinit=initial_value)

# Function to be called when the slider value changes
def update(val):
    # Get the current slider value
    value = slider.val
    # Update the y-values based on the slider value
    new_y = [val * y_val for y_val in y]
    line.set_ydata(new_y)
    fig.canvas.draw_idle()

# Register the update function with the slider
slider.on_changed(update)

# Show the plot and slider
plt.show()
```
